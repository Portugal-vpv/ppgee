```python
def create_training_datasets(function, lower_boundary, upper_boundary):
    datasets = []
    for i in range(20):
        x =  np.sort(np.random.uniform(lower_boundary, upper_boundary, 40))
        y = []
        for j in range(40):
            y.append(function(x[j]))
        datasets.append([x, y])
    return datasets
```

```python
       
def create_test_output(lower_bound, upper_bound, sample_ammount, function_name):
    x_test = np.linspace(lower_bound, upper_bound, sample_ammount)
    test_output = []
    for i in range(len(x_test)):
        test_output.append(function_name(x_test[i]))
    return x_test , test_output

def create_expected_values(training_method, dataset, lower_bound, upper_bound):
    x_test = np.linspace(lower_bound, upper_bound, 1000).reshape(-1, 1)
    y_pred = []
    expected_y =[]
    for i in range(len(dataset)):
        model = training_method(dataset[i][0], dataset[i][1])
        y_pred_new = model.predict(x_test)
        y_pred.append(y_pred_new)
        
    for j in range(len(y_pred[0])):
        y_result = []
        for k in range(len(y_pred)):
            y_result.append(y_pred[k][j])
        expected_y.append(np.mean(y_result))
        
    return y_pred, expected_y

def create_expected_values_for_constant_method(dataset, lower_bound, upper_bound):
    y_pred = []
    expected_y = []
    for i in range(len(dataset)):
        dataset_y = np.mean(dataset[i])
        y_pred.append(np.full(1000, dataset_y))
        expected_y.append(dataset_y)
        
    return y_pred, np.full(1000, np.mean(expected_y))

def create_expected_values_for_interpolation_methods(training_method, dataset, lower_bound, upper_bound):
    x_test = np.linspace(lower_bound, upper_bound, 1000)
    y_pred = []
    expected_y =[]
    for i in range(len(dataset)):
        model = training_method(dataset[i][0], dataset[i][1])
        y_pred_new = BSpline(*model)(x)
        y_pred.append(y_pred_new)
        
    for j in range(len(y_pred[0])):
        y_result = []
        for k in range(len(y_pred)):
            y_result.append(y_pred[k][j])
        expected_y.append(np.mean(y_result))
    
    return y_pred, expected_y

def calculate_metrics(y_pred, expected_y, y, y_true):
    mse = []
    bias = []
    varian = []
    
    for i in range(len(y_pred)):
        for j in range(len(y_pred[i])):
            mse.append((y[j] - y_pred[i][j])**2)
            bias.append((y_true[j] - expected_y[j])**2)
            varian.append((y_pred[i][j] - expected_y[j])**2)
              
    return np.mean(mse), np.mean(bias), np.mean(varian)
```

```python
def f1(x):
    #Domain: [-8 <= x <= 12]
    return (x-2)*(2*x +1)/(1 + x**2)

def f2(x):
    #Domain: [0 <= x <= 2pi]
    return math.sin(x)

def f3(x):
    #Domain: [0 <= x <= 10]
    return math.sin(2*x)/math.e **(x/5)

def f1_training(x):
    #Domain: [-8 <= x <= 12]
    return (x-2)*(2*x +1)/(1 + x**2) + error(0.3)

def f2_training(x):
    #Domain: [0 <= x <= 2pi]
    return math.sin(x) + error(0.3)

def f3_training(x):
    #Domain: [0 <= x <= 10]
    return math.sin(2*x)/math.e **(x/5) + error(0.3)

def error(std_deviation):
    return np.random.normal(loc=0, scale=std_deviation)
```

```python
def plot_graphs(title, x_label, y_label, y_1, y_2, y_3):
    x = [0, 2, 4, 6]
    plt.plot(x, y_1)
    plt.plot(x, y_2)
    plt.plot(x, y_3)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
    
plot_graphs("MSE", "grau de flexibilidade", "MSE", [mse_1_1, mse_2_1, mse_3_1, mse_4_1], [mse_1_2, mse_2_2, mse_3_2, mse_4_2], [mse_1_3, mse_2_3, mse_3_3, mse_4_3])
plot_graphs("Bias", "grau de flexibilidade", "Bias", [bias_1_1, bias_2_1, bias_3_1, bias_4_1], [bias_1_2, bias_2_2, bias_3_2, bias_4_2], [bias_1_3, bias_2_3, bias_3_3, bias_4_3])
```
